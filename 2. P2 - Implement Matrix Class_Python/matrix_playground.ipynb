```python
m1 = Matrix([
    [1, 2],
    [3, 4]
])

m2 = Matrix([
    [2, 5],
    [6, 1]
])

m3 = m1 + m2
print("m1 + m2 is")
print(m3)

m_determinant=Matrix.determinant(m1)
print("The determinant of this matrix: ")
print(m1)
print("is ", m_determinant)

m_trace=Matrix.trace(m1)
print("The trace of this matrix: ")
print(m1)
print("is ", m_trace)

m_inverse=Matrix.inverse(m1)
print("The inverse of this matrix: ")
print(m1)
print("is ", m_inverse)

m_trans=Matrix.T(m1)
print("The transpose of this matrix: ")
print(m1)
print("is ", m_trans)

m_addition=Matrix.__add__(m1,m2)
print("The addition of the matrix: ")
print(m1, "and", m2)
print("is ", m_addition)

m_neg=Matrix.__neg__(m1)
print("The negative of this matrix: ")
print(m1)
print("is ", m_neg)

m_sub=Matrix.__sub__(m1,m2)
print("The subtraction of the matrix: ")
print(m1, "and", m2)
print("is ", m_sub)

m_mul=Matrix.__mul__(m1,m2)
print("The multiplication of the matrix: ")
print(m1, "and", m2)
print("is ", m_mul)

m_rmul=Matrix.__rmul__(m1,2)
print("The multiple of this matrix: ")
print(m1, "and", 2)
print("is ", m_rmul)
```
